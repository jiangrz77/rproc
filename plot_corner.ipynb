```python
from astropy.io import fits
file =  fits.open('ba_sig5.fits')
data = file[0].data
import corner
title = ['Const', r'$\theta_\mathrm{age}$', r'$\theta_\mathrm{feh}$', r'$\theta_{corr}$', r'$\theta_{rg}$', r'$\theta_\sigma$']
corner.corner(
    data, 
    titles=title, smooth=.1,
    show_titles=True, plot_datapoints=True, 
    quatiles=[.16, .5 ,.84]
);
```
